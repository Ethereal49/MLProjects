```python
# 导入所需的库
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from matplotlib.font_manager import FontProperties

# 设置中文显示字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 'SimHei' 是黑体的字体名称
plt.rcParams['axes.unicode_minus'] = False  # 解决负号'-'显示为方块的问题
```

```python
# 假设的数据：房屋面积和价格
# 面积（平方米）
X = np.array([50, 60, 70, 80, 90, 100, 110, 120, 130, 140]).reshape(-1, 1)
# 价格（万元）
y = np.array([300, 350, 400, 450, 500, 550, 600, 650, 700, 750])
```

```python
# 创建线性回归模型
model = LinearRegression()
# 训练模型
model.fit(X, y)
```

```python
# 进行预测
X_test = np.array([150, 160]).reshape(-1, 1) # 测试数据
predictions = model.predict(X_test)

# 显示结果
print("预测的价格（万元）:", predictions)
```

```python
# 可视化
plt.scatter(X, y, color='blue') # 原始数据点
plt.plot(X, model.predict(X), color='red') # 预测的线性关系
plt.title('房屋面积与价格关系')
plt.xlabel('面积（平方米）')
plt.ylabel('价格（万元）')
plt.show()
```
